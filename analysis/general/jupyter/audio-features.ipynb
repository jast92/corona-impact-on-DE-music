```python
artisttweaked = [a.split(" feat. ")[0].split(" x ")[0].replace(" & ", " ").replace(" / ", " ").replace("'", "").replace(" [DE]", "").replace(".", " ").replace("?", "").replace("¿", "").replace(" +", "").replace(" [UK]", "").replace(" [2019]", "").replace(" [Live]", "") for a in basicinfo['artist']]
titletweaked = [t.split(" feat. ")[0].split(" x ")[0].replace(" & ", " ").replace(" / ", " ").replace("'", " ").replace(" [DE]", "").replace(".", " ").replace("?", "").replace("¿", "").replace(" +", "").replace(" [UK]", "").replace(" [2019]", "").replace(" [Live]", "") for t in basicinfo['title']]

searchterms = []
for a, t in zip(artisttweaked, titletweaked):
    s = a + " " + t
    searchterms.append(s)
```

```python
searchterms
```

```python
uris = []
for searchquery in tqdm(searchterms):
    try:
        usersearch = sp.search(q=searchquery, market="DE", limit=1)
        uri = usersearch["tracks"]["items"][0]["uri"]
    except:
        uri = np.nan
    uris.append(uri)
```
