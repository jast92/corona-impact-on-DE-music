### Create dataframe with audio features of songs using Spotify API

#### Set up access to Spotify API

```python
clientid = getpass.getpass('Enter Client ID: ')
clientsecret = getpass.getpass('Enter Client Secret: ')
```

```python
#Initialize SpotiPy with user credentials
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(
    client_id=clientid,
    client_secret=clientsecret))
```

```python
#Quick test - Searching songs with 'queries' with sp.search

results = sp.search(q='mein herz brennt rammstein', limit=50)
results['tracks']
```
