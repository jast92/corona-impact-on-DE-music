```python
songinfo = ['artist', 'title', 'label', 'year', 'month', 'week']
#y2018 = ['y' + str(2018) + '_' + str(week) for week in range(1,53)]
#y2019 = ['y' + str(2019) + '_' + str(week) for week in range(1,53)]
#y2020 = ['y' + str(2020) + '_' + str(week) for week in range(1,53)]

col_names = songinfo # + y2018 + y2019 + y2020

chartsall = pd.DataFrame(columns=col_names)
chartsall
```

```python
charts_2018_1 = pd.read_pickle("../pickle/charts_2018_1.pkl")
charts_2018_1
```

```python
chartsall['artist'] = charts_2018_1['artist']
chartsall['title'] = charts_2018_1['title']
chartsall['label'] = charts_2018_1['label']
chartsall['year'] = charts_2018_1['year']
chartsall['month'] = charts_2018_1['month']
chartsall['y2018_1'] = charts_2018_1['current_rank']
chartsall['week'] = 1
```

```python
charts_2018_2 = pd.read_pickle("../pickle/charts_2018_2.pkl")
charts_2018_2
```

```python
chartsall = chartsall.append(charts_2018_2[['artist', 'title', 'label']], ignore_index=True, sort=False)
```

```python
chartsall = chartsall.drop_duplicates(subset=['artist', 'title', 'label'], keep='first', ignore_index=True)
```

```python
chartsall['year'] = chartsall['year'].fillna(2018).astype(int)
```

```python
chartsall['month'] = chartsall['month'].fillna(1).astype(int)
```

```python
chartsall['week'] = chartsall['week'].fillna(2).astype(int)
```

```python
chartsall = pd.merge(chartsall, charts_2018_2[['artist', 'title', 'current_rank']], how='left', on=['artist', 'title'],
                     sort=False, validate="one_to_one")
```

```python
chartsall = chartsall.rename(columns={"current_rank": "y2018_2"})
```
